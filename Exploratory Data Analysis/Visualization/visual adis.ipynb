```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris_df = pd.read_csv("iris.csv")
iris_df.head(100)
```

```python
iris_df.describe()
```

```python

X = iris_df["SepalLengthCm"]
y = iris_df["Species"] 
z = iris_df["SepalWidthCm"] 
```

line chart

```python
plt.plot(X, y, marker='o', linestyle='-')
plt.title("Simple Line Chart")
plt.xlabel("X")
plt.ylabel("y")
```

bar chart

```python
plt.figure(figsize=(8, 6))
plt.bar(y,X, color='skyblue')
plt.title("Simple Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
```

SCATTER PLOT

```python
plt.scatter(y, X, color='blue', marker='d', label='Scatter Plot')
plt.title("Simple Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
```

POLAR CHART

```python

plt.figure(figsize=(8, 8))
plt.polar(y, X, marker='o', color='skyblue')
plt.fill(y, X, 'skyblue', alpha=0.4)
plt.title("Simple Polar Chart")
```

AXES PLOT & STACKED PLOT

```python
plt.figure(figsize=(8, 5))
plt.stackplot(y, X, alpha=0.5)
plt.title("Simple Stack Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
```

PIE CHART

```python
labels = ['Iris-setosa', 'Iris-virginica', 'Iris-versicolor']
tota=iris_df["Species"].value_counts()
r1 = iris_df[iris_df["Species"] =="Iris-setosa" ]
r2 = iris_df[iris_df["Species"] =="Iris-virginica"]
r3 = iris_df[iris_df["Species"] =="Iris-versicolor"]
c1 = r1["Species"].value_counts()
c2 = r2["Species"].value_counts()
c3 = r3["Species"].value_counts()
p1 = (c1 / sum(tota)) * 100
p2 = (c2 / sum(tota)) * 100
p3 = (c3 / sum(tota)) * 100
sizes = [p1,p2,p3] 
siz=pd.Series(sizes)
plt.figure(figsize=(6, 6))  # Set the figure size
plt.pie(siz, labels=labels, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightcoral', 'lightgreen', 'lightsalmon'])
plt.title("Simple Pie Chart")


```

HISTOGRAM

```python
plt.figure(figsize=(8, 6))
plt.hist(X, bins=15, color='skyblue', edgecolor='black')
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
```

TABLE CHART

```python
subset_df = iris_df.head(15)
fig, ax = plt.subplots(figsize=(8, 4))
ax.axis('off')
table = ax.table(cellText=subset_df.values,
                 colLabels=subset_df.columns,
                 cellLoc='center',
                 loc='center')
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1.2, 1.2)
plt.title("Iris Dataset Subset")
plt.show()
```

LOLIPOP CHART

```python
r1 = iris_df[iris_df["Species"] =="Iris-setosa" ]
r2 = r1["SepalLengthCm"]
index = range(1, len(r1) + 1)
plt.figure(figsize=(8, 5))
plt.stem(index, r2, linefmt='-b', markerfmt='ob', basefmt=' ', use_line_collection=True)
plt.title("Lollipop Chart ")
plt.xlabel("Data Point")
plt.ylabel("SepalLengthCm")
```
