```python
median_value = df['Sales of shampoo over a three year period'].median()
df['Sales of shampoo over a three year period'].fillna(median_value, inplace=True)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
# Decompose the time series
x=df['Sales of shampoo over a three year period'] 
decomposition= seasonal_decompose(x, model='additive', period=3)
# Plot the decomposed components
plt.figure(figsize=(10, 8))
plt.subplot(4, 1, 1)
plt.plot(df.index, df[ 'Sales of shampoo over a three year period'], label='Original')
plt.legend (loc='best')
plt.subplot(4, 1, 2)
plt.plot(df.index, decomposition.trend, label='Trend')
plt.legend (loc='best')
plt.subplot(4, 1, 3)
plt.plot(df.index, decomposition.seasonal, label='Seasonality')
plt.legend(loc='best')
plt.subplot(4, 1, 4)
plt.plot(df.index, decomposition. resid, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
#Calculate rolling statistics
rolling_mean =df[ 'Sales of shampoo over a three year period'].rolling (window=7).mean()
rolling_std=df['Sales of shampoo over a three year period'].rolling (window=7).std()
# Plot rolling statistics
plt.figure(figsize=(12, 6)) 
plt.plot(df.index, df['Sales of shampoo over a three year period'], label='Original') 
plt.title('Rolling Mean & Standard Deviation')
plt.plot(df.index, rolling_mean, label='Rolling Mean')
plt.plot(df.index, rolling_std, label='Rolling Std')
plt.xlabel('Month')
plt.ylabel('Sales of shampoo over a three year period')
plt. legend (loc='best')
plt.show()
```

```python
#Moving Average Smoothing
df['ma smooth']=df['Sales of shampoo over a three year period'].rolling (window=7).mean()
#Exponential Smoothing
df['exp_smooth'] = df['Sales of shampoo over a three year period'].ewm(span=7, adjust=False).mean()
# Plot the original and smoothed time series
plt.figure(figsize=(12, 6))
plt.plot(df. index, df['Sales of shampoo over a three year period'], label='Original')
plt.plot(df. index, df['ma smooth'], label='Moving Average')
plt.plot(df.index, df['exp_smooth'], label='Exponential Smoothing')
plt.title('Smoothing Techniques')
plt.xlabel('Month')
plt.ylabel('Sales of shampoo over a three year period')
plt.legend(loc='best')
plt.show()
```
